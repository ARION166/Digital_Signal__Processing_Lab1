```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define input vector x_mu and length N
x_mu = np.array([6, 8, 2, 4, 3, 4, 5, 0, 0, 0])
N = len(x_mu)  # N = 10

# 2. Build matrix K
# Use vectors k (column) and mu (row)
k_vec = np.arange(N).reshape(-1, 1)
mu_vec = np.arange(N).reshape(1, -1)
K = k_vec @ mu_vec

print("Matrix K (10x10):")
print(K)
print("-" * 50)

# 3. Build Fourier matrix W
# Note the sign convention: e^(+j * (2*pi/N) * K)
W = np.exp(1j * (2 * np.pi / N) * K)

print("Matrix W (10x10), rounded to 3 decimals:")
print(np.round(W, 3))
print("-" * 50)

# 4. Synthesize signal x_k using IDFT
# Formula: x_k = (1/N) * W * x_mu
x_k = (1 / N) * (W @ x_mu)

# Since x_mu does not have Hermitian symmetry,
# the synthesized signal x_k has both real and imaginary parts.
print("Synthesized signal x_k (complex values):")
print(np.round(x_k, 3))

# 5. Plot the signal
plt.figure(figsize=(10, 5))
plt.plot(np.real(x_k), marker='o', linestyle='-', label='Real part')
plt.plot(np.imag(x_k), marker='x', linestyle='--', label='Imaginary part')
plt.title('Synthesized signal x[k] (IDFT) - Variant 7')
plt.xlabel('Sample index k')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```
